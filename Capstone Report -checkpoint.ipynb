```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# pulling in the .csv file with the data 
data = pd.read_csv('C:/Users/Nolan/Desktop/Avocado.csv')
```

# Load and view data set 

```python
# making a data frame, and seeing the different variables available
avocado = pd.DataFrame(data)
avocado.head()
```



### This is a kaggle data set downloaded from the Hass Avocado Board website. They describe their data as weekly retail scan data for volume and price, coming directly from retailer's cash registers on actual sales of Hass Avocados. 

```python
avocado['Date']=pd.to_datetime(avocado['Date'])
```

#### Two of our data attributes are Average Price and Total Volume. If we make a scatter plot, we can investigate to see if these have any visible correlation between the two variables. 

```python
plt.scatter(x = avocado['AveragePrice'], y = avocado['Total Volume'],color = 'm',s=5,marker = '*')
plt.title('Average Price vs. Total Avocado Volume')
plt.xlabel('Average Price')
plt.ylabel("Total Avocado Volume")
plt.show()
```

### Although we do have a lot of outliers, it does appear from our plot that as price increases, the total volume becomes lower and lower. 

#### Let's investigate our data a little further for any unique attributes 

```python
avocado.type.unique()
```

```python
avocado.region.unique()
```

#### Based on the above, it looks like we have a data set that contains both organic and conventional avocado's, as well as various regions across the US from which the data came from. 

### Let's take another look at Price and Volume. How has Price and Volume moved over time for the dates in our data set?

```python
avocado[['Date','AveragePrice']].groupby('Date').mean().plot(color = 'red', figsize = (14,4))
plt.ylabel('Price Per Avocado')
plt.title('Average price over time')

avocado[['Date', 'Total Volume']].groupby('Date').mean().plot(color = 'purple',figsize = (14,4))
plt.ylabel('Avocados sold')
plt.title('Average total volume over time')
```

#### When we're looking at the peaks and valleys of the chart above, it certainly seems on the surface that as volume goes down prices tend to go up, and vice versa. 

#### Since Organic and conventional avocado's are categorized differently, and we have volume and average price data for both from across the United States, we can start to ask our self some interesting questions. 

# 1) Is there a difference in price between conventional and organic Avocado's?

#### Earlier when we were digging around in our data we saw two different type's of avocado being tracked. Conventional and organic. If we break our data set out into Organic vs Conventional, we can investigate whether or not there's a noticeable price difference between them. 

```python
avocado.groupby('type')[['AveragePrice']].describe()
```

#### Based on a bit of summary statistics it would certainly seem like there is a considerable difference in average price between the two. 

#### Let's visualize it. 

```python
conventional = avocado[avocado['type'] == 'conventional']
organic = avocado[avocado['type'] == 'organic']

c = plt.hist(conventional['AveragePrice'],bins = 20,alpha = .5, color = 'red')
o = plt.hist(organic['AveragePrice'],bins = 20, alpha = .5, color = 'blue')
plt.title('Organic vs. Conventional Price')
plt.xlabel('Average Price')
plt.legend('c''o')
```

#### This is a histogram of the average price between both conventional and organic avocado's. This is a pretty good visual because we can see which price most of the data falls around in regards to both conventional and organic. I think it shows pretty clearly that the bulk of organic data falls visibly higher than the conventional data. 

#### Based on the idea that Avocado prices are influenced in someways by the total volume, let's compare total volume of organic vs conventional avocado's to see if there is a discrepancy between the two. 

# 2) Is there a difference in volume between organic and conventional avocado's?

```python
avocado.groupby('type')[['Total Volume']].describe()
```

#### The average volume is much higher for conventional. We can also see the standard deviation for conventional is much larger than organic. The volume for conventional avocado's must have a lot more dispersion in its data compared to organic. 

#### Let's look further into aggregate levels between the two. 

```python
avocado.groupby('type')[['AveragePrice']].mean().plot(kind='bar')
plt.title('Conventional vs Organic Average Price')
plt.ylabel('Price Per Avocado')

avocado.groupby('type')[['Total Volume']].mean().plot(kind = 'bar',color = 'red')
plt.title('Conventional vs Organic Total Volume')
plt.ylabel('Total Avocados Sold')
```

#### These box plots give us a possible reason why there is a difference in price between organic and conventional avocado's. When you compare total volume, organic gets dwarfed by the volume of conventional. The difference in volume between the two makes some sense on the surface. Organic vegetables are more expensive and tedious to produce. Not all farmers then are willing to convert their land to organic cultivation, especially when a majority of the consumers are happy with conventional avocado's. 

```python
avocado[['AveragePrice','type']].boxplot(by='type',figsize = (15,10))
```

#### Again, we see here that organic has a higher average price than conventional. It also looks like organic has a lot more outliers in price on both ends of the spectrum, while conventional only seems to have price outliers at the top. 

```python
avocado[['Total Volume','type']].boxplot(by='type',figsize = (15,5))
```

#### This again shows how large, and how many, outliers there are when you're looking at total volume for conventional avocado's. This isn't a good visual because of how wide spread our volume data is for conventional avocado's compared to organic. It's hard to compare the averages. 

#### Based on the visuals above, as well as some summary statistics, I think there is a clear difference in price between organic and conventional avocado's and that a portion of that price difference can be attributed to the difference in total volume between the two.  



## Are Avocado prices going to be different based on regions? 



#### From earlier we know there are quite a few unique regions in our data set. The regions span from small to large cities, regions covering multiple states, and the whole US itself. If we want to see an effect on avocado price based on the region of the country, we can take cities that are geographically spread out across the country, and have populations that aren't always consistent. 




```python
avocado_by_region = avocado[avocado['region'].isin(['Boston','Chicago','Columbus','NewYork','Nashville','SanFrancisco','LosAngeles'])]
```

```python
avocado_by_region[['AveragePrice','region']].boxplot(by = 'region',figsize = (15,8))
plt.ylabel('Price per Avocado')
plt.xlabel('Cities')
```

## We can see it looks like the average price fluctuations are based more on city size, than geographical region. Of the larger cities, Los Angeles is the only one with an average price in line with a smaller city.  

## Let's take a look at the volume for each of these cities to see if there is more to the story. 

```python
avocado_by_region[['Total Volume','region']].boxplot(by = 'region', figsize = (15,8))
plt.ylabel("Total Avocado's")
plt.xlabel('Cities')
```

### Los Angeles has an extremely high volume of Avocado's, which falls in line with it having a lower price. However, Nashville and Columbus seem to have a much smaller volume, yet their prices were visibly lower than the other major cities. I think this shows that in certain instances location does have an effect on prices that aren't always connected to total volume. Cost of living and what people are willing to pay probably comes into factor at a certain point as well.
