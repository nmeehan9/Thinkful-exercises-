```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('C:/Users/Nolan/Desktop/Avocado.csv')
```

```python
df.info()
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
organic_df = df[df['type']=='organic']
conventional_df = df[df['type']=='conventional']
```

```python
organic_df[['Date','AveragePrice',]].groupby('Date').mean().plot()
conventional_df[['Date','AveragePrice',]].groupby('Date').mean().plot()
```

```python
avg_price_region = conventional_df.groupby('region')[['AveragePrice']].mean().sort_values(by=['AveragePrice'])
avg_vol_region = df.groupby('region')[['Total Volume']].mean().sort_values(by = ['Total Volume'])
```

```python
avg_price_region.plot(kind = 'barh',figsize = (12,16))
plt.title('Average Price per Region')
```

```python
avg_vol_region.plot(kind = 'barh', figsize = (12,16))
```

```python

df[['Date','AveragePrice',]].groupby('Date').mean().plot()


df[['Date','Total Volume']].groupby('Date').mean().plot()
```

It shows here that as the Volume increases, Average Price with decrease, and vice versa. 

```python
organic_conventional = df.groupby('type')[['AveragePrice']].mean()
```

```python
organic_conventional.plot(kind = 'bar', figsize=(10,6))
```
